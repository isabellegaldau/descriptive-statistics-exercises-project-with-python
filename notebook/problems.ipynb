### Exercise 1

We will use Numpy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO
import numpy as np
import statistics as stats
from scipy.stats import skew
from scipy.stats import kurtosis

normal_data = np.random.normal(0, 1, 100)
chi_data = np.random.chisquare(3, 100)

# mean
n_mean = np.mean(normal_data)
c_mean = np.mean(chi_data)

# meadian
n_median = np.median(normal_data)
c_median = np.median(chi_data)

# mode
n_mode = stats.multimode(normal_data)
c_mode = stats.multimode(chi_data)

# range 
n_range = np.max(normal_data) - np.min(normal_data)
c_range = np.max(chi_data) - np.min(chi_data)

# varaiance
n_var = np.var(normal_data)
c_var = np.var(chi_data)

# standard deviation 
n_std = stats.stdev(normal_data)
c_std = stats.stdev(chi_data)

# skew
n_skew = skew(normal_data)
c_skew = skew (chi_data)

# kurtosis
n_kurt = kurtosis(normal_data)
c_kurt = kurtosis(chi_data)

print (f'Normal Distribution Metrics \n Mean:{n_mean}\n Median:{n_median} \n Mode:{n_mode} \n Range:{n_range} \n Varaiance:{n_var} \n Std:{n_std} \n Skew:{n_skew} \n Kurtosis:{n_kurt}')
print (f'Chi Distribution Metrics \n Mean:{c_mean} \n Median:{c_median} \n Mode:{c_mode} \n Range:{c_range} \n Varaiance:{c_var} \n Std:{c_std} \n Skew:{c_skew} \n Kurtosis:{c_kurt}')
#print (agg, n_mean, n_median)
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
data = [4, 2, 5, 8, 6]
mean = sum(data) / len(data)
deviations = [x - mean for x in data]
squared_deviations = [x**2 for x in deviations]
mean_squared_deviations = sum(squared_deviations) / len(deviations)
standard_dev = mean_squared_deviations**0.5
print("Standard Deviation:", standard_dev)
````
